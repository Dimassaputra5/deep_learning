```python
from torchvision import transforms
from luwiji.cnn import illustration
```

# Data Augmentation

augmentasi: Sedikit mengubah data asli
- meningkatkan generality (mengurangi overfit)<br>
- menambah data<br>

**It is a best practice to perform data augmentation when possible.**

```python
img = illustration.img
img
```

### Random Crop

```python
crop = transforms.RandomCrop(256)
crop(img)
```

### Random Flip

```python
hflip = transforms.RandomHorizontalFlip()
hflip(img)
```

```python
vflip = transforms.RandomVerticalFlip()
vflip(img)
```

### Random Rotation

```python
rotate = transforms.RandomRotation(10)
rotate(img)
```

### Random ResizedCrop

```python
resize = transforms.RandomResizedCrop(256)
resize(img)
```

### Random Affine

```python
translate = transforms.RandomAffine(0, translate=(0.3, 0.3))
translate(img)
```

```python
shear = transforms.RandomAffine(0, shear=10)
shear(img)
```

```python
affine = transforms.RandomAffine(5, translate=(0.1, 0.1), shear=10)
affine(img)
```

### Color Jitter

```python
brightness = transforms.ColorJitter(brightness=0.3)
brightness(img)
```

```python
contrast = transforms.ColorJitter(contrast=0.3)
contrast(img)
```
